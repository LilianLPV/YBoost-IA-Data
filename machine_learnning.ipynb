```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler, LabelEncoder 
```

```python
df1 = pd.read_csv('meteo_juillet_decembre_2019.csv')
df2 = pd.read_csv('Travel_titles_with_departments.csv')
```

```python
fusion = pd.merge(df1.rename(columns={'JOUR': 'DATE'}), df2, left_on='DATE', right_on='DATE' , how='inner')[lambda x: x['DEPARTEMENT'] == 75].reset_index(drop=True)
fusion["DATE"] = pd.to_datetime(fusion["DATE"])
```

```python
fusion.head()
```

```python
fusion.info()
```

```python
X = fusion.drop(["NB_VALID", "DATE"], axis=1) 
Y = fusion["NB_VALID"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression(max_iter=1000) 
```

```python
model.fit(X_train, y_train) 
```

```python
y_pred = model.predict(X_test) 
```

```python
y_prob = model.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 
print("Accuracy:", accuracy_score(y_test, y_pred)) 
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred)) 
print(classification_report(y_test, y_pred)) 
```

```python
from sklearn.metrics import mean_squared_error, r2_score 
print("MSE:", mean_squared_error(y_test, y_pred)) 
print("R2:", r2_score(y_test, y_pred)) 
```
